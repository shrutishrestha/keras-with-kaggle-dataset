```python
import keras
import pandas
import pandas as pd
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers.core import Activation,Dense,Flatten
from keras.layers.convolutional import Conv2D
from keras.layers.core import Activation,Dense,Flatten
from keras.layers.pooling import MaxPool2D
from sklearn.model_selection import train_test_split
```

```python
train_data = pd.read_csv('/home/shruti/python/AISaturdays/Kaggle-MNIST/train.csv')
test_data = pd.read_csv('/home/shruti/python/AISaturdays/Kaggle-MNIST/test.csv')
```

```python
 features = train_data.iloc[:,1:].values.reshape(len(train_data),28,28,1)
```

```python
labels = train_data.iloc[:,0].values
```

```python
labels.shape
```

```python
labels = keras.utils.to_categorical(labels,num_classes=10)
```

```python
labels.shape
```

```python
x_train,x_test,y_train,y_test = train_test_split(features,labels,test_size=0.20,random_state=1)
#train-test-validation
#test is kaggle ko data
#atako test is validation
```

creating our model

```python
def LeNet(length,width,output,channels):
    model = Sequential()
    #convolutional
    model.add(Conv2D(filters=32,kernel_size=(2,2),strides=(1,1),input_shape=(length,width,channels)))
    
    #Activation
    model.add(Activation('relu'))
    
    #Pooling
    model.add(MaxPool2D(pool_size=(2,2)))
    
    #convolutional
    model.add(Conv2D(filters=64,kernel_size=(2,2),strides=(1,1)))
    
    #Activation
    model.add(Activation('relu'))
    
    #Pooling
    model.add(MaxPool2D(pool_size=(2,2)))
    
    model.add(Flatten())
    
    model.add(Dense(100))
    
    model.add(Activation('relu'))
    
    model.add(Dense(output))
    
    model.add(Activation('softmax'))
    
    return model
```

```python
model = LeNet(28,28,10,1)
```

```python
model.summary()
```

```python
model.compile(loss="categorical_crossentropy",optimizer="adadelta",metrics=['accuracy'])
```

```python
epochs=30
```

```python
model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=epochs)
```

```python
model.save_weights('mnist.h5')
```
